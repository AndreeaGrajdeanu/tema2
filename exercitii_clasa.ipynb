```python
x=np.linspace(10,80,20)
random_sample=np.random.randint(10,80,1000)
threshes = np.dstack((x[:-1], x[1:])).reshape(x.size - 1, 2)
print(threshes)
```

```python
counts = np.array([random_sample[(random_sample >= t[0]) & (random_sample < t[1])].size / random_sample.size for t in threshes])
print(counts)
```

```python
plt.plot(x[:-1], counts, label = 'density_function')

plt.xlabel('x')
plt.ylabel('Count')

plt.show()
```
