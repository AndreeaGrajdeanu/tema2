#### Step 3. Assign it to a variable called chipo

```python
db=ps.sqldf("select sum(quantity) as sum_Q from chipo;", locals())
db
```

#### Step 5. What is the price of each item?

```python
ps.sqldf("select item_name, item_price from chipo;", locals())
```

#### Step 6. Sort by the name of the item

```python
ps.sqldf("select *  from chipo order by item_name;", locals())
```

```python
ps.sqldf("select *  from chipo order by quantity;", locals())
```

#### Step 7. What was the quantity of the most expensive item ordered?

```python
data_new=ps.sqldf("select item_price, REPLACE(item_price,'$','') from chipo;", locals())
data_new
```

```python
data_new.columns
```

```python
ps.sqldf("select MAX(REPLACE(item_price,'$','')) from data_new;", locals())
```

```python
oredr_data=ps.sqldf("select * from data_new order by REPLACE(item_price,'$','') DESC;", locals())
oredr_data
```

```python
new_new=ps.sqldf("select CAST(REPLACE(item_price,'$','') as FLOAT) from data_new;", locals())
new_new
```

```python
new_new.columns
```

```python
ps.sqldf("select MAX(CAST(REPLACE(item_price,'$','') as FLOAT)) from data_new;", locals())
```

#### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
salad=ps.sqldf("select COUNT(item_name) from chipo where item_name='Veggie Salad Bowl' ;", locals())
salad
```

#### Step 9. How many times people orderd more than one Canned Soda?

```python
canned=ps.sqldf("select COUNT(item_name) from chipo where (item_name='Canned Soda') and (quantity>1) ;", locals())
canned
```
